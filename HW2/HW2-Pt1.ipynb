```python
# 3. Find/print the latitude and longitude of all airlines in the ‘United States’. (Join Airlines and Airports tables)


query = "SELECT L.name,latitude,longitude\
         FROM airports as P JOIN airlines as L \
         ON P.icao=L.icao WHERE L.country='United States';"

team_list = pandas.read_sql(query, conn)
team_list.head()
```

```python
#4. Find/print the average latitude for each country listed in the Airports table.
query = "SELECT country,AVG(latitude) From airports GROUP BY country;"

team_list = pandas.read_sql(query, conn)
team_list.head()
```

```python
#6. Find/print all the routes originating from airports in the city of ‘Los Angeles’. (use Routes and Airports tables
query = "SELECT *\
         FROM airports as A JOIN routes as R \
         ON A.code=R.source\
         WHERE city= 'Los Angeles';"

team_list = pandas.read_sql(query, conn)
team_list.head()
```
