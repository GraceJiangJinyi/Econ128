```python
df=dataI.join(dataZ.set_index('Zip'), on='Zip')
df.head()
```

```python
df.shape
```

```python
df.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
#drop lines with null value

df = df.dropna(subset = ['State']) 
df = df.dropna(subset = ['County']) 
df = df.dropna(subset = ['City']) 
df = df.dropna(subset = ['ZrentValue']) 
print 'State null is', sum(df['State'].isnull())
print 'County null is', sum(df['County'].isnull())
print 'City null is', sum(df['City'].isnull())
print 'ZrentValue null is', sum(df['ZrentValue'].isnull())
```

```python
df.set_index('Zip', inplace = True)
df.head()
```

```python
import math
df["NoReturns"] = df["NoReturns"].apply(lambda x: math.log(x+1))
df["SingleReturns"] = df["SingleReturns"].apply(lambda x: math.log(x+1))
df["UnempComp"] = df["UnempComp"].apply(lambda x: math.log(x+1))
df["StudentLoanDeductions"] = df["StudentLoanDeductions"].apply(lambda x: math.log(x+1))
df["TaxableIncome"] = df["TaxableIncome"].apply(lambda x: math.log(x+1))
df["ZrentValue"] = df["ZrentValue"].apply(lambda x: math.log(x+1))
df.head()
```

```python
df_reg = pd.DataFrame(df[['NoReturns', 'SingleReturns', 'TaxableIncome','StudentLoanDeductions', 'UnempComp','ZrentValue']])
df_reg.head()


```

```python
y = pd.DataFrame(df_reg[['ZrentValue']])
x = pd.DataFrame(df_reg[['NoReturns', 'SingleReturns', 'TaxableIncome','StudentLoanDeductions', 'UnempComp']])

x.head()
```

```python
#fit the OLS regression
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
model = lm.fit(x,y)
model
```

```python
print('estimated intercept:', lm.intercept_)
print('estimated coefficients:', lm.coef_)
print('features:',list(x.columns.values))
```

```python
import statsmodels.api as sm

x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
model.summary()
```

```python
print "A 1% increase in the number of returns field would increase the expected Rent Value by",lm.coef_[0,0] , '%'
print "A 1% increase in the number of single returns field would increase the expected Rent Value by",lm.coef_[0,1], '%'
print "A 1% increase in the total taxable income would decrease the expected Rent Value by",lm.coef_[0,2]*(-1),'%'
print "A 1% increase in the total student loan interest deductions would decrease the expected Rent Value by",lm.coef_[0,3]*(-1), '%'
print "A 1% increase in the total unemployment compensation would increase the expected Rent Value by",lm.coef_[0,4],'%'
```
