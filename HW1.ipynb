```python
#numerical data: age, income, children

print "Min_Age: " , min(df['age'])
print "Max_Age: " , max(df['age'])
print "Std_Age: " , df['age'].std()
print "Avg_Age: " , df['age'].mean()
print""
print "Min_Income: " , min(df['income'])
print "Max_Income: " , max(df['income'])
print "Std_Income: " , df['income'].std()
print "Avg_Income: " , df['income'].mean()
print""
print "Min_Children: " , min(df['children'])
print "Max_Children: " , max(df['children'])
print "Std_Children: " , df['children'].std()
print "Avg_Children: " , df['children'].mean()
```
